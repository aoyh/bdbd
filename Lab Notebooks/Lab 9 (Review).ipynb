### Study Tip
I would recommend thinking through the following list of questions for each of the key methods (RCT, IV, RD and DID):
1. What is a description of of the methodology? (also be able to identify the method from a description)
2. How are the treatment and control group defined?  How is the outcome defined?
3. What conditions do you need to satisfy to ensure you can estimate an unbiased treatment effect (make causal link between treatment and control)?  What empirical tests can you run to ensure these conditions are satisfied?
